```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(font_scale=1.5)

# To open the csv's with the results
main_path = '/home/rl3155/Multilingual-Speech-Emotion-Recognition-System/Spanish/Spanish_New'
wav2vecbase = pd.read_csv(main_path + '/results/wav2vecbase_perlayer.csv')
wav2veclarge = pd.read_csv(main_path + '/results/wav2veclarge_perlayer.csv')
hubertbase = pd.read_csv(main_path + '/results/hubertbase_perlayer.csv')
hubertlarge = pd.read_csv(main_path + '/results/hubertlarge_perlayer.csv')

# Calculate the overall y-limits for the data
y_min = min(wav2vecbase['accuracy'].min(), hubertbase['accuracy'].min(),
            wav2veclarge['accuracy'].min(), hubertlarge['accuracy'].min()) - 0.03
y_max = min(1, int( (max(wav2vecbase['accuracy'].max(), hubertbase['accuracy'].max(),
            wav2veclarge['accuracy'].max(), hubertlarge['accuracy'].max()) + 0.11)*100)/100)

# Create a figure with two subplots arranged side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(19, 7), dpi = 100)

# Plot the accuracy for wav2vecbase and hubertbase in the first subplot
ax1.plot(range(12), wav2vecbase['accuracy'], linewidth=3, label='wav2vecbase')
ax1.plot(range(12), hubertbase['accuracy'], linewidth=3, label='hubertbase')
ax1.set_xlabel('Layer')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.set_title('Base models (12 layers)')
ax1.set_ylim([y_min, y_max])


# Plot the accuracy for wav2veclarge and hubertlarge in the second subplot
ax2.plot(range(24), wav2veclarge['accuracy'], linewidth=3, label='wav2veclarge')
ax2.plot(range(24), hubertlarge['accuracy'], linewidth=3, label='hubertlarge')
ax2.set_xlabel('Layer')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.set_title('Large models (24 layers)')
ax2.set_ylim([y_min, y_max])

fig.suptitle("Spanish -- Performance of Different Feature Extractors per Layer")

plt.show()
```
