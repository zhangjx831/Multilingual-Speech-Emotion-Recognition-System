```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

layers = 24

main_path = '../emotiondata/'
wav2vecbase = pd.read_csv(main_path + '/results/wav2vecbase_perlayer.csv')
wav2veclarge = pd.read_csv(main_path + '/results/wav2veclarge_perlayer.csv')
hubertbase = pd.read_csv(main_path + '/results/hubertbase_perlayer.csv')
hubertlarge = pd.read_csv(main_path + '/results/hubertlarge_perlayer.csv')

plt.plot(range(12), wav2vecbase['accuracy'], label = 'wav2vecbase')
plt.plot(range(12), hubertbase['accuracy'], label = 'hubertbase')
plt.plot(range(24), wav2veclarge['accuracy'], label = 'wav2veclarge')
plt.plot(range(24), hubertlarge['accuracy'], label = 'hubertlarge')

plt.xlabel("Layer")
plt.ylabel("Accuracy")
plt.legend()
plt.axis([0, layers-1, 0, 1])
plt.xticks(np.arange(0, layers, 1))
plt.title("Different Feature Extractors Per Layer Performance")
plt.show()
```
