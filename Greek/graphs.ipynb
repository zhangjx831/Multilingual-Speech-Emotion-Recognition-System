```python
import pandas as pd

layers = 12
df = pd.DataFrame()
for layer in range(layers):
    file = f'emotiondata/results/hubertbase_perclass_{layer}.csv'
    data = pd.read_csv(file)
    data['layer'] = layer
    df = pd.concat([data, df], ignore_index=True)
df['accuracy'] = df['accuracy'].round(2)
df.head()
```

```python
newdf = pd.pivot(df, index='layer', columns='class', values='accuracy')
newdf
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(newdf, annot=True, cmap="crest")
plt.title('Greek HubertBase Confusion Matrix in Layers')
```
