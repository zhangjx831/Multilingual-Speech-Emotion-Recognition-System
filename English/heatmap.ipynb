```python
import pandas as pd
layers = 12
df = pd.DataFrame()

for layer in range(layers):
    file = f'/home/xg2378/results/hubertbase_perclass_{layer}.csv '
    data = pd.read_csv(file)
    data['layer'] = layer
    df = pd.concat([data, df], ignore_index=True)
df['accuracy'] = df['accuracy'].round(2)
df.head()
     
```

```python

newdf = pd.pivot(df, index='layer', columns='class', values='accuracy')
newdf
    
```

```python
import seaborn as sns

sns.heatmap(newdf, annot=True, cmap="crest").set(title="English: HubertBase Each Class Accuracy Via Layers")
     
```
